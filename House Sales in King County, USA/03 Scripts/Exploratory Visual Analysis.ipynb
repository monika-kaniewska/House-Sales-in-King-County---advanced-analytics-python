#### Correlation Insights

**Price**

There is a strong positive correlation between price and sqft of living area, sqft above, grade, and the number of bathrooms. These factors indicate that they are the main drivers of price. Additionally, there is moderate positive correlation between price and number of bedrooms, view and square feet of the basement. This implies that while these features do contribute to the price, their impact is not as strong as that of the core factors above. Lastly, there is a weak positive correlation between price and number of floors, waterfront and year renovated. This suggests that while these features may slightly affect the price, their influence is relatively minor compared to other variables.

**Bedrooms**

There is a strong positive correlation between bedrooms and bathrooms, as well as between bedrooms and square feet of living area. This means that, the more bedrooms a house has, the larger the living area tends to be, and the more bathrooms are typically included. Moreover, there is also a moderate positive relationship between bedrooms and price, grade, sqft above and sqft basement. This can imply that houses with more bedrooms are generally larger, better designed, and have higher in prices.

**Bathrooms**

There is a strong positive correlation between bathrooms and price, bedrooms, sqft living area, number of floors, grade, sqft_above, year built, and the average neighbors living area, which means that houses with more bathrooms tend to be larger, better built, and generally more valuable.

**Sqft_living**

Theres a strong positive correlation between square feet of living area and price, bedrooms, bathrooms, grade, sqft_above, and neighbors living area. This means that larger homes are more expensive, have more rooms, and are placed in better neighborhoods. The number of floors, view, sqft of basement, and year built are linked to size moderately.

**Sqft_land_area** 

Sqft_land_area has a generally weak relationship with most variables, but it shows a strong positive correlation with the average land area of the neighbors, which means that homes with larger land areas tend to be in neighborhoods where other properties have also lager land areas.

**Floors**

The number of floors has a moderate to strong positive correlation with the number of bathrooms, sqft_living, grade,  sqft_above, and year built. This indicates, that homes with larger amount of floors tend to have more bathrooms, larger living spaces, better quality, and are often newer.

**Waterfront** 

Waterfront has a moderate correlation with view, indicating that waterfront properties are often associated with better views.

**Grade**

Grade strongly correlates with price, bathrooms, SQFT living, SQFT above, and neighbors living area, and moderately with bedrooms, floors, and year built. Surprisingly, it has a weak negative correlation with condition, which means that higher-grade homes arent always in better condition.

**Year Built**

Year Built positively correlates with bathrooms, sqft_living, grade, and floors. It has a negative correlation with sqft_basement, waterfront and view, which sugegst that newer homes often lack basements and value space over view.

The scatterplot reveals the positive relationship between the number of bathrooms and price. The points are mostly clusterred up to 5 bahrooms, after which, the price is influenced by some high value outliers and a few low value data points.

```python
# Creating the scatterplot of price and bedrooms
sns.lmplot(x='bedrooms', y='price', data=df)
```

Although the correlation coefficient between price and the number of bedrooms reveals a moderate positive relationship of 0.32, there is a bell curve distribution of bedrooms in relation to price. The scatterplot reveals that the price increases with the grade.

#### Categorical Plot

```python
# Creating histogram to visualize the distribution of price
sns.histplot(df['price'], bins = 20, kde = True)
plt.show()
```

```python
df.describe()
```

```python
# Defining price categories
```

```python
# Low price category - up to the first quartile
df.loc[df['price'] < 324844, 'Price category'] = 'Low price'
```

```python
# Middle price category - between Q1 and Q3
df.loc[(df['price'] >= 324844) & (df['price'] < 645000 ), 'Price category'] = 'Middle price'
```

```python
# High price category - more than Q3
df.loc[df['price'] >= 645000, 'Price category'] = 'High price'
```

```python
# Checking the value counts
df['Price category'].value_counts()
```

```python
df.columns
```

```python
# Creatting a categorical plot using the price categories and view
sns.set(style="ticks")
g = sns.catplot(x="view", y="price", hue="Price category", data=df, palette={'Low price': '#1f77b4', 'Middle price': '#ff7f0e', 'High price': '#2ca02c'})
plt.show()
```

The majority of low-priced houses have a view quality of 0 or 2. Interestingly, the chart also shows that it is possible to find houses with a very good view (4) at lower prices. Conversely, there are many high-priced houses with a view rating of 0, which may suggest that buyers value other property features more than the view itself.

```python
# Creating a categorical plot using the price categories and number of bedrooms
sns.set(style="ticks")
g = sns.catplot(x="bedrooms", y="price", hue="Price category", data=df, palette={'Low price': '#1f77b4', 'Middle price': '#ff7f0e', 'High price': '#2ca02c'})
plt.show()
```

Low-price houses tend to have up to 7 bedrooms. The number of bedrooms for middle-priced houses is widely distributed across the chart, with the highest density between 1 and 7 bedrooms. For high priced houses, the number of bedrooms varies from 0 to 10, but the majority falls between 3 and 5 bedrooms, which suggests that extremely large houses (7+ bedrooms) are uncommon even among expensive properties.

```python
# Create a categorical plot using the price categories created above
sns.set(style="ticks")
g = sns.catplot(x="condition", y="price", hue="Price category", data=df, palette={'Low price': '#1f77b4', 'Middle price': '#ff7f0e', 'High price': '#2ca02c'})
plt.show()
```

There are only a few high-priced houses in poor condition (1 or 2). The majority of high-priced houses have a condition rating between 3 and 5. Surprisingly, some low-priced houses are offered in very good condition at a relatively low price.

```python
# Create a categorical plot using the price categories created above
sns.set(style="ticks")
g = sns.catplot(x="grade", y="price", hue="Price category", data=df, palette={'Low price': '#1f77b4', 'Middle price': '#ff7f0e', 'High price': '#2ca02c'})
plt.show()
```

Low-priced houses tend to have lower grades, and their number is very limited. Middle-priced houses generally have grades between 4 and 11, which indicates that most middle price houses have average to above-average quality. High-priced houses tend to have much higher grades, and their prices also vary significantly. This means that house grade is an important factor influencing price.
